```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import os
```

```python
os.getcwd()
```

```python
df=pd.read_csv("E:\Machine Learning\Home_Prices.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

Predict price of a home with area = 3300 sqr ft


```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
