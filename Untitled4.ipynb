```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.ebay.com/p/228576584?iid=254473643400"
r=requests.get(url)
htmlcontent=r.content
```

```python
soup=BeautifulSoup(htmlcontent,'html.parser')
```

```python
print(soup.prettify())
```

```python
print(soup.find('title').get_text())
print(soup.find(class_='product-title').get_text())
print(soup.find(class_='price').get_text())
```
