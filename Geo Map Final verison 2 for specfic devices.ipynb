```python
from bs4 import BeautifulSoup
import requests
import urllib
```

```python
visitedlinks=set()
```

```python
def find_all_device_links(Url):
    if Url=='#':
        return
    try:
        device_links=[]
        htmlContent = urllib.request.urlopen(Url).read()
        soup=BeautifulSoup(htmlContent,'html.parser')    
        div_data=soup.find('div',attrs={'class':'goods_'})
        links=div_data.find_all('a')
        for i in links:
            print(i['href'])
            device_links.append(i['href'])
    except:
        print("error in all_device_links")
    return device_links
```

```python
def get_second_page_data(Url,device_name,grade,price ,price2):
    if Url=='#':
        return
    try:
    
        htmlContent = urllib.request.urlopen(Url).read()
        soup=BeautifulSoup(htmlContent,'html.parser')
        second_div_data=soup.find('div',attrs={'class':"goodsspec_"})
        if(second_div_data==None):
            return
        g_obj=second_div_data.find('span',attrs={'class':"icon1_"})
        if(g_obj==None):
              return
        g=g_obj.find('img',alt=True)['alt']


        d_obj=second_div_data.find('h1',attrs={'class':'goods_name_'})
        if(d_obj==None):
              return
        d=d_obj.getText()

        p_obj=second_div_data.find('span',attrs={'class':'goods_detail_price_'})
        if(p_obj==None):
              return
        p=p_obj.text.replace('\r','').replace('\n',"").strip()

        p2_obj=second_div_data.find('span',attrs={'class':'goods_detail_price_small_'})
        if(p2_obj==None):
            return
        p2=p2_obj.text.replace('\r','').replace('\n',"").strip()

        grade.append(g)
        device_name.append(d)
        price.append(p)
        price2.append(p2)
        print(g,d,p,p2,Url)
    except:
        print("error in second page")
```

```python
def pagechange(Url):
    if Url ==None:
        return '#'
    try:
        pagelinks=[]
        htmlContent = urllib.request.urlopen(Url).read()
        soup=BeautifulSoup(htmlContent,'html.parser')
        navigation_data=soup.find('div',attrs={'class':'navipage_ bottom_'})
        allpagelinks=navigation_data.findAll('a')
        #print(allpagelinks)
        for i in allpagelinks:
            print('https://ec.geo-online.co.jp'+i['href'])
            pagelinks.append('https://ec.geo-online.co.jp'+i['href'])
    except:
        print("errorn in page change")
        
    return pagelinks
    
```

```python
device_name=[]
grade=[]
price=[]
price2=[]
pagelinks=list()
urllinks=[]
pageindex=0;
```

```python

for url in urllinks:
    pagelinks.append(url)
    one=True
    for page in pagelinks:
        pageindex+=1
        print("page number",pageindex,page)
        if(page not in visitedlinks and page!='#'):
            visitedlinks.add(page)
        else:
            continue

        devicelinks=find_all_device_links(page)
        for deviceUrl in devicelinks:
            if(deviceUrl not in visitedlinks and deviceUrl !='#'):
                visitedlinks.add(deviceUrl)
            else:        
                continue
            get_second_page_data(deviceUrl,device_name,grade,price,price2)

        if one:
            temppagelinks=pagechange(page)
            ind_temp=0
            for link in temppagelinks:
                if link not in pagelinks and link !='#':
                        pagelinks.append(link)
                        ind_temp+=1 
                if ind_temp==1:
                    break
            one=False
```
