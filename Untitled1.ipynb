```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://ec.geo-online.co.jp/shop/goods/search.aspx?search.x=0&keyword=&goods_code=&store=&tree=15501009&genre_tree=&capacity=&price=&flg="
r=requests.get(url)
```

```python
htmlContent =r.content
print(htmlContent)
```

```python
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup)
```

```python
print(soup.prettify)
```

```python
title=soup.title
print(title)
```

```python
print(title.string)
```

```python
all_links= set()
anchors=soup.find_all('a')
#print(anchors)
for link in anchors:
    if(link.get('href')!="#"):
        all_links.add(link.get('href'))
print(all_links)
```
