```python
def find_all_links(soup):
    links_set=set()
    anchors=soup.find_all('a')
    for links in anchors:
        if(links.get('href')!="#"):
             links_set.add(links.get('href'))
    return links_set
```

```python
links_set=find_all_links(soup)
```
