```python
from bs4 import BeautifulSoup
import requests
import urllib
```

```python
Url="https://www.phonearena.com/phones/manufacturers/apple"
```

```python
htmlContent = urllib.request.urlopen(Url).read()
soup=BeautifulSoup(htmlContent,'html.parser')    
```

```python
print(soup)
```

```python
div_data=soup.find('div',attrs={'class':'container-fluid phones-listing'})
links=div_data.find_all('a')
```

```python
device_links=set()
for link in links:
    if(link.get('href')!='#'):
        device_links.add(link.get('href'))
    
```
