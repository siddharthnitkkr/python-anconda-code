```python
from bs4 import BeautifulSoup
import requests
import urllib
```

```python
Url="https://ec.geo-online.co.jp/shop/goods/search.aspx?search.x=0&keyword=&goods_code=&store=&tree=1510&genre_tree=&capacity=&price=&flg="
```

```python
#
#r=requests.get(Url)
#htmlContent=r.content
htmlContent = urllib.request.urlopen(Url).read()


#with urllib.request.urlopen("http://www.python.org") as url:
 #   s = url.read()
    # I'm guessing this would output the html source code ?
  #  print(s)
```

```python
#htmlContent = urllib.urlopen(Url).read()
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify())
```

```python
div_data=soup.find('div',attrs={'class':'goods_'})
links=div_data.find_all('a')
for i in links:
    print(i['href'])
```

```python
second_link="https://ec.geo-online.co.jp/shop/g/g2300140031161/"
htmlContent = urllib.request.urlopen(second_link).read()
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify())
```

```python
second_div_data=soup.find('div',attrs={'class':"goodsspec_"})
second_div_data.find('span',attrs={'class':"icon1_"}).find('img',alt=True)['alt']
second_div_data.find('h1',attrs={'class':'goods_name_'}).getText()
```

```python
second_div_data.find('span',attrs={'class':'goods_detail_price_'}).text.replace('\r','').replace('\n',"").strip()
```

```python
#page change
navigation_data=soup.find('div',attrs={'class':'navipage_ bottom_'})
allpagelinks=navigation_data.findAll('a')
#print(allpagelinks)
for i in allpagelinks:
    print('https://ec.geo-online.co.jp'+i['href'])


```
