```python
from bs4 import BeautifulSoup
import requests
```

```python
def set_soup(url):        
    r=requests.get(url);
    htmlContent=r.content
    soup=BeautifulSoup(htmlContent,'html.parser')
    return soup
```

```python
def find_all_links(soup):
    links_set=set()
    anchors=soup.find_all('a')
    for links in anchors:
        if(links.get('href')!="#"):
             links_set.add(links.get('href'))
    return links_set
```

```python
def find_all_data(global_soup,links_set,device,price,visited_links):
    for link in links_set:
        if("www.ebay.com" not in link):
            continue
        if(link in visited_links):
            continue
        else:
            visited_links.add(link)
        if(("pgn=") in link):
            local_soup=set_soup(link)
            local_links_set=find_all_links(local_soup)
            find_all_data(local_soup,local_links_set,device,price,visited_links)
        else:
            local_soup=set_soup(link)
            dev_obj=local_soup.find(class_='product-title')            
            pri_obj=local_soup.find(class_ ='price')
            if(dev_obj==None and pri_obj==None):
                print(link,"Both None")
                
            elif(dev_obj!=None and pri_obj==None):
                 print(link,"Second None")
            elif(dev_obj==None and pri_obj!=None):
                 print(link,"first None")
            else:
                dev=dev_obj.get_text()
                pri=pri_obj.get_text()               
                
                print(link , dev , pri)
                device.append(dev)
                price.append(pri)
                
            
            
        
```

```python
#main
url="https://www.ebay.com/b/Apple-Cell-Phones-Smartphones/9355/bn_319682"
visted_links=set()
soup=set_soup(url)
link_set=find_all_links(soup)
device=list()
price=list()
find_all_data(soup,link_set,device,price,visted_links)

```
